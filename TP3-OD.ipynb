```python
import pandas as pd

# Étape 0 : Garder uniquement les données de température (et non l'écart-type)
df = df[df["Element"] == "Temperature change"]

# Étape 1 : Sélection des colonnes correspondant aux années
year_columns = [col for col in df.columns if col.startswith("Y")]

# Étape 2 : Transformation du DataFrame en format long
df_long = pd.melt(
    df,
    id_vars=["Area", "Months"],
    value_vars=year_columns,
    var_name="Year",
    value_name="Temperature change"
)

# Étape 3 : Nettoyage de la colonne "Year"
df_long["Year"] = df_long["Year"].str[1:].astype(int)

# Étape 4 : Ordre réel des mois
mois_ordre = ["January", "February", "March", "April", "May", "June", 
              "July", "August", "September", "October", "November", "December"]
df_long["Months"] = pd.Categorical(df_long["Months"], categories=mois_ordre, ordered=True)

# Étape 5 : Tri du DataFrame
df_long = df_long.sort_values(by=["Area", "Year", "Months"])

# Étape 6 : Moyenne mobile sur 5 ans
df_long["RollingMean_5Y"] = df_long.groupby(["Area", "Months"])["Temperature change"].transform(
    lambda x: x.rolling(window=5, min_periods=1).mean()
)

# Étape 7 : Aperçu
print(df_long.head(20))
```

```python
# Supprimer les lignes contenant des NaN dans n'importe quelle colonne
df_long_clean = df_long.dropna()

# Vérifier le résultat
print(df_long_clean.head(20))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 📊 1. Evolution des anomalies de température pour une année donnée
def plot_temperature_evolution_for_year(df_long_clean, year):
    # Filtrer les données pour l'année spécifiée
    df_year = df_long_clean[df_long_clean["Year"] == year]
    
    # Créer un graphique de l'évolution de la température par mois
    plt.figure(figsize=(10, 6))
    sns.lineplot(x="Months", y="Temperature change", data=df_year, marker='o', color="b")
    plt.title(f"Evolution des anomalies de température en {year} pour chaque mois")
    plt.xlabel("Mois")
    plt.ylabel("Anomalie de température (°C)")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# 📊 2. Evolution des anomalies de température avec la moyenne mobile sur 5 ans pour une année donnée
def plot_temperature_with_rolling_mean(df_long_clean, year):
    # Filtrer les données pour l'année spécifiée
    df_year = df_long_clean[df_long_clean["Year"] == year]
    
    # Créer un graphique de l'évolution de la température et de la moyenne mobile
    plt.figure(figsize=(10, 6))
    sns.lineplot(x="Months", y="Temperature change", data=df_year, marker='o', label="Anomalie de température", color="b")
    sns.lineplot(x="Months", y="RollingMean_5Y", data=df_year, marker='o', label="Moyenne mobile (5 ans)", color="r", linestyle="--")
    plt.title(f"Anomalies de température et moyenne mobile sur 5 ans en {year}")
    plt.xlabel("Mois")
    plt.ylabel("Anomalie de température (°C)")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.show()

# 📊 3. Evolution des anomalies de température pour une période de plusieurs années (par exemple, 1969-2019)
def plot_temperature_evolution(df_long_clean, start_year, end_year):
    # Filtrer les données pour les années spécifiées
    df_years = df_long_clean[(df_long_clean["Year"] >= start_year) & (df_long_clean["Year"] <= end_year)]
    
    # Créer un graphique de l'évolution des anomalies de température
    plt.figure(figsize=(12, 6))
    sns.lineplot(x="Year", y="Temperature change", data=df_years, hue="Months", marker='o', palette="coolwarm", lw=2)
    plt.title(f"Evolution des anomalies de température de {start_year} à {end_year}")
    plt.xlabel("Année")
    plt.ylabel("Anomalie de température (°C)")
    plt.legend(title="Mois", loc="upper left")
    plt.grid(True)
    plt.show()

# 📊 Utiliser les fonctions
# Exemple : Analyser l'année 1970
plot_temperature_evolution_for_year(df_long_clean, 1970)

# Exemple : Analyser la température et la moyenne mobile pour 1970
plot_temperature_with_rolling_mean(df_long_clean, 1970)

# Exemple : Analyser la période de 1969 à 2019
plot_temperature_evolution(df_long_clean, 1969, 2019)
```

phase 3: Visualisation spatiale et géographique

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcul de la température moyenne annuelle par mois
df_long['Year'] = df_long['Year'].astype(str)
avg_temp_per_month = df_long.groupby(['Months', 'Year'])['Temperature change'].mean().unstack()

# Tracer un graphique des températures moyennes par année
plt.figure(figsize=(10,6))
avg_temp_per_month.T.plot(kind='line', marker='o', figsize=(12, 8))
plt.title('Température moyenne annuelle par mois (1969-2019)')
plt.xlabel('Mois')
plt.ylabel('Température moyenne (°C)')
plt.xticks(rotation=45)
plt.legend(title='Année', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Tracer la moyenne mobile sur 5 ans
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_long, x='Year', y='RollingMean_5Y', hue='Months', marker='o')
plt.title('Moyenne mobile sur 5 ans des températures (1969-2019)')
plt.xlabel('Année')
plt.ylabel('Température moyenne mobile (°C)')
plt.xticks(rotation=45)
plt.legend(title='Mois', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Tracer un histogramme des températures par mois
plt.figure(figsize=(12, 6))
sns.histplot(df_long['Temperature change'], bins=30, kde=True, color='pink')

plt.title('Distribution des températures en France (1969-2019)')
plt.xlabel('Température (°C)')
plt.ylabel("Nombre d'occurrences")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calcul de l'écart-type des températures par mois sur toutes les années
std_temp_by_month = df_long.groupby('Months')['Temperature change'].std()

# Tracer l'écart-type des températures par mois
plt.figure(figsize=(12, 6))
sns.barplot(x=std_temp_by_month.index, y=std_temp_by_month.values, palette='viridis')
plt.title("Variabilité interannuelle des températures (1969-2019)")
plt.xlabel('Mois')
plt.ylabel("Écart-type des températures (°C)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
